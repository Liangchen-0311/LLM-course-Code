## BestPractice


### 基本流程

- 明确目标：效果、速度、并发等定量指标。
- 设计整体架构：考虑资源和目标的平衡。
- 评估效果：不同模块独立评估，可以使用不同LLM。
- 编写程序：进入开发，充分考虑不同系统和可能的扩展，比如RAG、管理后台、提示词配置等。模块间尽量解耦。充分考虑可追溯。充分考虑二次优化。
- 整体评估：全流程评估效果，根据结果优化代码，打通全流程。
- 持续优化

### 注意事项

- 根据需要设计，但同时预留一定的扩展性。
- 模块功能和业务逻辑分离。
- 过程追溯（包括耗时记录）要持久化。
- Human-in-the-Loop很重要，应与自动训练/微调、部署关联起来。
- 多方面评估。
- 不拘泥形式（比如意图分类代替LLM）。
